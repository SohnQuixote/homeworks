```python
lis = open('list.txt' , mode = 'r' , encoding = 'utf-8')
links = lis.readlines()
```

```python
links
```

```python
search = "\n"
for i, word in enumerate(links):
    if search in word: 
        print('>> modify: ' + word)
        links[i] = word.strip(search)
```

```python
links
```

```python
result = []
for i in range(len(links)):
    url = 'https://m.blog.naver.com/hwasinedu/'+str(links[i])
    resp = requests.get(url)
    soup = BeautifulSoup(resp.content)
    p = soup.find_all('a' , class_ = "se-file-save-button __se_link" )
    for a in p:
        result.append(a.attrs['href'])
```

```python
result
```

```python
for i in range(0 , len(result)):
    url = result[i]
    res = requests.get(url)
    with open(str(i)+ '.pdf' , 'wb') as f:
        f.write(res.content)
```
